### Visualização, análise e sumarização dos dados

Análise de preços por país e categoria

```python
#Classificação por Categoria e país
dash1=cplp.groupby(['Categoria', 'País'])['Preço (USD)'].sum().reset_index(name='Preço Global').sort_values(by='Preço Global', ascending =False)
```

```python
dash1
```

```python
# configurando as cores para cada país
cores=[]
for i in dash1['País']:
    if i =='Angola':
        cores.append('red')
    elif i == 'Brasil':
        cores.append('yellow')
    elif i == 'Cabo Verde':
        cores.append('blue')
    elif i == 'Moçambique':
        cores.append('green')
    else:
        cores.append('darkred')


#configurando o gráfico
mpl.figure(figsize =(16,6))
sb.barplot(x='Categoria', y='Preço Global', hue='País', data=dash1, palette=cores).set(title='Distribuição de Preços por Categoria e País')
mpl.xticks(rotation = 25)
mpl.show()
```

Análise global de preços por categoria

```python
dash2=cplp.groupby(['Categoria'])['Preço (USD)'].sum().reset_index(name='Preços').sort_values(by='Preços', ascending =False)
```

```python
dash2
```

```python
#configurando o gráfico
dsh2plot=sb.barplot(y='Categoria', x='Preços', data=dash2, palette='tab10')
dsh2plot.bar_label(dsh2plot.containers[0])
sb.set(rc={'figure.figsize':(16,6)}, font_scale=1)
sb.set_style({'axes.facecolor':'white', 'grid.color': '.5', 'font.family':'Times New Roman'})
dsh2plot.set(title='Preços por Categoria');
```

### Considerações finais

Sobre os preços na CPLP:
    1. A média de salários em Angola é mais alta, o que amortiza, em certa medida, o acesso a produtos e serviços. Tal situação, no entanto, não se verifica em Moçambique que, muito próximo dos números globais de Angola, é o país com a média de salários mais baixa da comunidade.
    2. Um outro ponto de realce é a relação entre o poder da moeda e o preço de bens e serviços. Para cada unidade de um bem ou serviço comprado em Kwanzas (AOA), é possível adquirir oito do mesmo em Meticais (MZN).

Sobre o OCR, pelo menos para o modelo em português (por.traineddata), foram verificadas algumas leituras imprecisas. Mas a margem de imprecisão foi baixa, tornando-a, por isso, num recurso imprescindível para extração de informações em imagens.
